```python
import pandas as pd
 
# Define the category mapping
category_mapping = {
    1: 'Down',
    2: 'No Booking Capability',
    3: 'Bounce Castle Network',
    4: 'Event Rental Systems',
    5: 'Inflatable Office',
    6: 'Event Office',
    7: 'Others'
}

# 1: Down
# 2: Booking Capability: Bounce Castle Network, Event Rental Systems, Inflatable Office, Event Office
# 3: No Booking Capability
# 4: Others
```

```python
new_df = pd.DataFrame(columns=list(df.keys())+["category_y", "category_name"])
```

```python
new_df
```

```python
i = 0
e = 0
for result in results:
    try:
        url = result['url']
        category = result['category']
        category_name = category_mapping[category]
        row = df[df['site'] == url]
        row['category_y'] = category
        row['category_name'] = category_name
        # print(row)
        new_df.loc[i] = row.values[0]
    except:
        print("Error: {}".format(e))
        e += 1
        pass
    i += 1
    # print(i)

# new_df
```
